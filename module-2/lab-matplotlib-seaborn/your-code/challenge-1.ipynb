#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
#import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1,2,1)
plt.plot(x,y, linewidth=2.0, color='r')
plt.title('y=2·x',size=12,fontweight='bold')  
plt.subplot(1,2,2)
plt.plot(x,z, linewidth=5.0, color='y')
plt.title('z=x^2',size=12,fontweight='bold')
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
for i in range(2):
    if i==0:
        f=y
        title='y=2·x'
        linewidth=2.0
        color='r'
    else:
        f=z
        title='z=x^2'
        linewidth=5.0
        color='y'
        
    plt.subplot(1,2,i+1)
    plt.plot(x,f, linewidth=linewidth, color=color)
    plt.title(title,size=12,fontweight='bold')  
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
plt.figure(figsize=(12,12))
for i in range(2):
    if i==0:
        f=y
        title='y=2·x'
        linewidth=2.0
        color='r'
    else:
        f=z
        title='z=x^2'
        linewidth=5.0
        color='y'
        
    plt.subplot(1,2,i+1)
    plt.plot(x,f, linewidth=linewidth, color=color)
    plt.title(title,size=12,fontweight='bold')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
plt.figure(figsize=(12,12))
for i in range(2):
    if i==0:
        f=y
        title='y=2·x'
        linewidth=2.0
        color='r'
        ylabel='f(x)'
    else:
        f=z
        title='z=x^2'
        linewidth=5.0
        color='y'
        ylabel=''
        
    plt.subplot(1,2,i+1)
    plt.plot(x,f, linewidth=linewidth, color=color)
    plt.ylabel(ylabel, size=12)
    plt.xlabel('X', size=12)
    plt.title(title,size=12,fontweight='bold')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
t=np.exp(x)
plt.figure(figsize=(12,12))
plt.plot(x,z, linewidth=2.0, color='r', label='x^2')
plt.plot(x,t, linewidth=2.0, color='b', label='exp(x)')
plt.ylabel('F(X)', size=12)
plt.xlabel('X', size=12)
plt.xscale('linear')
plt.yscale('log')
plt.title('x^2 vs exp(x)',size=12,fontweight='bold') 
plt.legend(loc='best')
```
